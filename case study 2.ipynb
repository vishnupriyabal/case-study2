# 2. student's parental level of education

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.subplots(figsize=(15,5))
data['parental level of education'].value_counts(normalize=True)
data['parental level of education'].value_counts(dropna=False).plot.bar(color=['yellow','green','black','blue','red'])
plt.title('Comparison of Parental Education')
plt.xlabel('Degree')
plt.ylabel('Count')
plt.show()
```
